Clean DFs by removing not required classes

```python
# =========================================
# STOCK_DF PROCESSING
# =========================================

# 1. Create a unique pair identifier combining Stock Name and Company Name
stock_df["pair"] = stock_df["Stock Name"] + "___" + stock_df["Company Name"]

# 2. Map each unique pair to an integer
pair_to_id = {pair: idx for idx, pair in enumerate(stock_df["pair"].unique())}

# 3. Create the numeric 'label' column
stock_df["label_clas"] = stock_df["pair"].map(pair_to_id)

# 4. Drop temporary and unnecessary columns
columns_to_drop = ["pair", "Stock Name", "Date", "Company Name"]
stock_df.drop(columns=columns_to_drop, inplace=True)

# 5. Rename 'Tweet' column to 'text' to match sentiment_df
stock_df.rename(columns={"Tweet": "text"}, inplace=True)
stock_df["text"] = stock_df["text"].astype(str)

# 6. Inspect
print("Stock DataFrame Head after processing:")
print(stock_df.head())
print("\nFinal Columns in stock_df:")
print(stock_df.columns)


# =========================================
# SENTIMENT_DF PROCESSING
# =========================================

# 1. Rename columns to match stock_df
sentiment_df.rename(columns={"Text": "text", "Sentiment": "label_sen"}, inplace=True)
sentiment_df["text"] = sentiment_df["text"].astype(str)

# 2. Inspect
print("\nSentiment DataFrame Head after processing:")
print(sentiment_df.head())
print("\nFinal Columns in sentiment_df:")
print(sentiment_df.columns)
```

From the below graph we can see that the data is very skewed

```python
# === STOCK_DF ===
# Count the number of records per company
company_counts = stock_df['label_clas'].value_counts()
company_ratios = company_counts / company_counts.sum()

# Plot company record distribution
plt.figure(figsize=(10, 6))
company_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Records per Company')
plt.xlabel('Company Name')
plt.ylabel('Number of Records')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Display ratios and counts
print("=== Company Record Ratios ===")
print(company_ratios)
print("\n=== Company Record Counts ===")
print(company_counts)


# === SENTIMENT_DF ===
# Count the sentiment distribution
sentiment_counts = sentiment_df['label_sen'].value_counts()
sentiment_ratios = sentiment_counts / sentiment_counts.sum()

# Plot sentiment distribution
plt.figure(figsize=(10, 6))
sentiment_counts.plot(kind='bar', color='lightcoral')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Number of Records')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# Display ratios and counts
print("\n=== Sentiment Ratios ===")
print(sentiment_ratios)
print("\n=== Sentiment Counts ===")
print(sentiment_counts)
```

```python


def make_balanced_sampler(df, label_col="label"):
    # Count the number of samples per class
    class_counts = df[label_col].value_counts().sort_index().tolist()
    num_classes = len(class_counts)

    # Compute weights for each class (inverse frequency)
    class_weights = [1.0 / count for count in class_counts]

    # Assign a weight to each sample in the dataset
    sample_weights = [class_weights[label] for label in df[label_col]]

    # Create a sampler
    sampler = WeightedRandomSampler(weights=sample_weights, num_samples=len(sample_weights), replacement=True)
    return sampler
    
def load_custom_dataframe(df, tokenizer, max_length=None):
    dataset = Dataset.from_pandas(df)

    if tokenizer.model_max_length > 1e5 or tokenizer.model_max_length <= 0:
        max_length = 512  # fallback to a safe max length
    else:
        max_length = tokenizer.model_max_length

    def preprocess_function(examples):
        return tokenizer(examples["text"], truncation=True, max_length=max_length)

    tokenized_dataset = dataset.map(preprocess_function, batched=True)
    return tokenized_dataset


def get_collator(tokenizer):
    """Returns a collator for padding batches"""
    return DataCollatorWithPadding(tokenizer=tokenizer)

def get_loaders_from_dataframe(train_df, test_df, batch_size, model_name,label_col="label"):
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    
    # Convert DataFrames to tokenized datasets
    train_dataset = load_custom_dataframe(train_df, tokenizer)
    test_dataset = load_custom_dataframe(test_df, tokenizer)

    data_collator = get_collator(tokenizer)

    train_dataset.set_format("torch", columns=["input_ids", "attention_mask", "label"])
    test_dataset.set_format("torch", columns=["input_ids", "attention_mask", "label"])

    # Create a WeightedRandomSampler for the training set
    train_sampler = make_balanced_sampler(train_df,label_col)

    train_loader = DataLoader(
        train_dataset, batch_size=batch_size, sampler=train_sampler, collate_fn=data_collator
    )
    test_loader = DataLoader(
        test_dataset, batch_size=batch_size, shuffle=False, collate_fn=data_collator
    )

    return train_loader, test_loader

```

```python
# Load pretrained sentiment model
model_name = "StephanAkkerman/FinTwitBERT-sentiment"
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Keep original sentiment head
sentiment_head = model.classifier  # pretrained weights

# Add a new parallel head for your classification task
num_new_classes = company_counts
num_sentiment_classes = 2
# new_head = nn.Linear(model.config.hidden_size, num_new_classes)

class HeadCNN(nn.Module):
    def __init__(self, new_num_classes, in_features=768):
        super().__init__()
        self.fc1 = nn.Linear(in_features, 512)
        self.relu = nn.ReLU(inplace=True)
        self.dropout = nn.Dropout(0.2)
        self.fc2 = nn.Linear(512, new_num_classes)
        self.final_act = nn.Identity()

    def forward(self, x):
        x = self.dropout(self.relu(self.fc1(x)))
        x = self.fc2(x)
        x = self.final_act(x)
        return x  # shape [B, new_num_classes]
        
class FrozenBERTMultiHead(nn.Module):
    def __init__(self, model_name, num_sentiment_classes, num_new_classes,freeze_head = False, freeze_param = True):
        super().__init__()
        # Load pretrained sentiment model
        self.bert_model = AutoModelForSequenceClassification.from_pretrained(
            model_name,
            num_labels = num_sentiment_classes  # change sentiment head output size
        )

        # Freeze BERT parameters
        if freeze_param:
            for param in self.bert_model.bert.parameters():
                param.requires_grad = False

        # Freeze sentiment head optionally
        if freeze_head:
            for param in self.bert_model.classifier.parameters():
                param.requires_grad = False

        # New classification head
        self.new_head = HeadCNN(in_features=self.bert_model.config.hidden_size,
                                new_num_classes=new_num_classes)

    def forward(self, input_ids, attention_mask):
        # Get shared BERT embeddings
        outputs = self.bert_model.bert(input_ids=input_ids, attention_mask=attention_mask)
        cls_embedding = outputs.last_hidden_state[:, 0, :]  # [CLS] token

        # Compute logits for both heads
        sentiment_logits = self.bert_model.classifier(cls_embedding)
        new_logits = self.new_head(cls_embedding)

        return sentiment_logits, new_logits
```
